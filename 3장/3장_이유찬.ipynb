```python
# 3.3.1

m = np.linspace(0,600,30).reshape((5,6))
print(m)
print(np.max(m))
print(np.sum(m,axis=1))
print(np.max(m,axis=1))
print(np.mean(m,axis=0))
print(np.min(m,axis=0))
```
