````python
# 2.6.4
def diffsum(a,b,c):
    s = a+b+c
    s2 = a**2+b**2+c**2
    if s>s2:
        return s-s2
    return s2-s
````

````python
# 2.6.5
m=0

def diffsum2(a,b,c):
    global m
    s = a+b+c
    s2 = a**2+b**2+c**2
    if s>s2:
        if s-s2 > m:
            m = s-s2
        return s-s2
    if s2-s > m:
        m = s2-s
    return s2-s
````

````python
# 2.12.1

class Triangle(object):
    def __init__(self, b,h):
        self.b = b
        self.h = h
    def area(self):
        return self.b*self.h*0.5
    
````

````python
# 2.12.2

class Cube(object):
    def __init__(self, a, b, h):
        self.a = a
        self.b = b
        self.h = h
    def surface(self):
        return 2*(self.a*self.b+self.b*self.h+self.a*self.h)
````

````python
# 2.12.3

class Character(object):

    def __init__(self):
        self.life = 1000
        self.strength = 10
        self.intelligence = 10

    def attacked(self):
        self.life -= 10
        print("공격받음! 생명력 =", self.life)

    def attack(self):
        print("공격!")

class Warrior(Character):

    def __init__(self):
        super(Warrior, self).__init__()
        self.strength = 15
        self.intelligence = 5

    def attack(self):
        print("육탄 공격!")
    
    def attacked(self):
        self.life -=1
        print("공격받음! 생명력 =", self.life)

class Wizard(Character):

    def __init__(self):
        super(Wizard, self).__init__()
        self.strength = 5
        self.intelligence = 15

    def attack(self):
        print("마법 공격!")
    
    def attacked(self):
        self.life-=2
        print("공격받음! 생명력 =", self.life)
````

````python
class Car(object):
    def __init__(self):
        self.max_speed = 160
        self.speed = 0
    def speed_up(self):
        if(self.max_speed > self.speed):
            self.speed+=20
        print(f"Speed : {self.speed}")
    def speed_down(self):
        if(self.speed >0):
            self.speed -=20
        print(f"Speed : {self.speed}")
a = Car()
a.speed_up()
a.speed_down()
a.speed_down()
a.speed_up()
a.speed_up()
for i in range(5):
    a.speed_down()
````

````python
# 2.12.5
class SportCar(Car):
    def __init__(self):
        super(SportCar, self).__init__()
        self.max_speed = 200
    def speed_up(self):
        if(self.max_speed-45 > self.speed):
            self.speed+=45
        print(f"Speed : {self.speed}")
    def speed_down(self):
        if(self.speed >0):
            self.speed -=45
        print(f"Speed : {self.speed}")


class Truck(Car):
    def __init__(self):
        super(Truck, self).__init__()
        self.max_speed = 100
    def speed_up(self):
        if(self.max_speed-15 > self.speed):
            self.speed+=15
        print(f"Speed : {self.speed}")
    def speed_down(self):
        if(self.speed >0):
            self.speed -=15
        print(f"Speed : {self.speed}")

a = SportCar()
a.speed_down()
a.speed_up()
a.speed_up()
a.speed_up()
a.speed_up()
a.speed_up()
a.speed_down()
a.speed_down()
for i in range(5):
    a.speed_down()

a = Truck()
a.speed_down()
a.speed_up()
a.speed_up()
a.speed_up()
a.speed_up()
a.speed_up()
a.speed_down()
a.speed_down()
for i in range(5):
    a.speed_down()
````

````python
# 2.12.6
class Student(object):
    def __init__(self,id,name,math,english):
        self.id = id
        self.name = name
        self.math=math
        self.english = english
    def __repr__(self):
        return f"이름 : {self.name} 학번 : {self.id}"
    def __str__(self):
        return self.name
    def __getitem__(self,key):
        if key == 'math':
            return self.math
        if key == 'english':
            return self.english
a = Student('ㅊ321043','NoName',10,20e20)
print(a)
print(a['math'], a['english'])
````
