```python
# 5.1.1

import matplotlib.pyplot as plt
import numpy as np
X = np.arange(1,11)
Y1 = np.random.randint(0,100,10)
Y2 = np.random.normal(0,1,10)*42.8+21.4
Y3 = np.random.random(10)*100
plt.plot(X,Y1,marker = 'o', c='r', ls="-", label = 'randint')
plt.plot(X,Y2 ,marker = "^", c = 'b', ls = '--', label = 'normal')
plt.plot(X,Y3, marker = 'x', c = 'g', ls = '-.', label = 'random')
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend(loc=2)
```

```python
# 5.1.2

fig, axs = plt.subplots(3,3,figsize=(10,10))
X = np.linspace(0,100,100)
START = 0
for i, j in [(x,y) for x in range(3) for y in range(3)]:
    Y = [START]
    for k in range(len(X)):
        Y.append(Y[-1]+np.random.normal()*(i+j/3))
    axs[i,j].plot(Y)
    axs[i,j].set_title(f"Cumulated Normal with {(i+j/3):.3} Standard Deviation")
    axs[i,j].set_xlabel("Time")
    axs[i,j].set_ylabel("Cumulation")
plt.tight_layout()
plt.show()
```

```python
# 5.1.3

N = 400
X = np.random.random(N)*50
# Y = X * 1.3 + 4 + (np.random.random(N)*2-1)*100
# Y = X * 1.3 + 4 + np.random.normal(0,10,N)
Y = np.log(X+1) + np.sin(X/4)*0.4+ np.random.normal(0,0.3,N)

a = sum((X-X.mean())*(Y-Y.mean()))/sum((X-X.mean())**2)
b = Y.mean() - a*X.mean()

err = a*X+b-Y
std =np.sqrt(sum(err**2)/(N-2))
print(std)
X_lin = np.linspace(X.min(),X.max(), 100)
Y_lin = a*X_lin+b


plt.scatter(X,Y)
plt.plot(X_lin, Y_lin, ls = "--", c='r')

a_s=[0.4,0.25,0.1]
for i in range(1,4):
    plt.fill_between(X_lin, Y_lin+std*i, Y_lin-std*i,alpha=a_s[i-1], linewidth=0,color='g')

plt.show()
```
